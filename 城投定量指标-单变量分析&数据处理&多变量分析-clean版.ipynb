## 1.2单变量基础分析

```python

"""

定义basi函数 计算最小值，最大值，平均值，零值个数，中位数，99%分位数，1%分位数，极端值个数，标准差，数据缺失率。

"""
def basic(x):   
    return pd.Series({
            'ObsNum': obs_num(x),
            'MIN': np.nanmin(x),
            'MAX': np.nanmax(x),
            'MEAN': np.nanmean(x),
            'Zero': zero_num(x),
            'Median' : np.nanmedian(x),
            'Percentile_99' : percentile_99(x),
            'Percentile_1' : percentile_1(x),
            'Extreme_num' :Extreme_num(x),
            'STD': np.nanstd(x),
            'Num': len(x),
            'MissRatio': miss_ratio(x),
            })


"""

计算 非空个数

"""
def obs_num(x):
    return len(x) - sum(np.isnan(x))


"""

计算 缺失率

"""
def miss_ratio(x):
    return sum(np.isnan(x)) / len(x)


"""

计算 零值个数

"""
def zero_num(x):
    
    return sum(x.apply(lambda x : 1 if x==0 else 0))


"""

计算 99%分位数

"""
def percentile_99(x):
    
    return np.percentile(a=x,q=99)

"""

计算 1%分位数

"""
def percentile_1(x):
    
    return np.percentile(a=x,q=1)

"""

计算 异常值

"""
def  Extreme_num(x):
    
    P_99 = np.percentile(a=x,q=99)
    
    P_1 = np.percentile(a=x,q=1)
    
    return sum(x.apply(lambda x : 0 if (x>=P_1)&(x<=P_99) else 1))
```

## 1.3均值及中位数分析

```python
"""

好坏样本集：按照好坏标识，分为好样本集和坏样本集

"""

#好样本集，仅包含指标列、排序和好坏标识
raw_data_good = raw_data[raw_data['bad_flag']==0].astype(float)

#del raw_data_good['年份']


#坏样本集，仅包含指标列、排序和好坏标识
raw_data_bad = raw_data[raw_data['bad_flag']==1].astype(float)

#del raw_data_bad['年份']
```

```python
"""

均值检验：确定各个指标的好样本均值和坏样本均值

"""

def mean_test(x,name):   
    return pd.Series({
            name : np.nanmean(x)
            })

mean_good = raw_data_good.iloc[:,:-1].apply(mean_test,args=('好样本均值',)).T

mean_bad = raw_data_bad.iloc[:,:-1].apply(mean_test,args=('坏样本均值',)).T


mean_final = pd.merge(left = mean_good, right = mean_bad, left_index=True,right_index=True)
mean_final
```

```python
"""

筛选通过均值检验的指标,假设均为正向指标，若为非正向指标，需要将好坏样本均值乘以-1

"""

indicator_mean_pass = []

for each in mean_final.index:
    
    if  mean_final.loc[each,'好样本均值'] >= mean_final.loc[each,'坏样本均值'] :
        
        indicator_mean_pass.append(each)
```

```python
"""

中位数检验：确定各个指标的好样本中位数和坏样本中位数

"""

def median_test(x,name):   
    return pd.Series({
            name : np.nanmedian(x)
            })

median_good = raw_data_good.iloc[:,:-1].apply(median_test,args=('好样本中位数',)).T

median_bad = raw_data_bad.iloc[:,:-1].apply(median_test,args=('坏样本中位数',)).T


median_final = pd.merge(left = median_good, right = median_bad, left_index=True,right_index=True)
median_final
```

```python
"""

筛选通过中位数检验的指标,假设均为正向指标，若为非正向指标，需要将好坏样本中位数乘以-1

"""

indicator_median_pass = []

for each in median_final.index:
    
    if  median_final.loc[each,'好样本中位数'] >= median_final.loc[each,'坏样本中位数'] :
        
        indicator_median_pass.append(each)
```

## 1.4缺失值处理

```python
"""

缺失值分组处理：好样本的缺失值等于好样本中位数，坏样本的缺失值等于坏样本中位数

"""

def nan_fix(x):
    
    med = np.nanmedian(x)
    
    x = x.fillna(med)
    
    
    return x



#对好样本集进行处理，补充缺失值
data_good_indicator = raw_data_good[indicator_pass_final].apply(nan_fix)

#对坏样本集进行处理，补充缺失值
data_bad_indicator = raw_data_bad[indicator_pass_final].apply(nan_fix)
```

## 1.5极端值处理

```python
"""

极端值统一处理：按照1%和99%分位数进行处理，小于1%的令其等于1%分位数，大于99%令其等于99%分位数

"""

def abnormal_fix(x):
    P_99 = np.percentile(a=x,q=99)
    
    P_1 = np.percentile(a=x,q=1)
    
    x = x.apply(lambda x : P_99 if x>=P_99 else x)
    
    x = x.apply(lambda x : P_1 if x<=P_1 else x)
    
    return x



#极端值处理，将好坏样本集合进行处理

data_good_indicator = data_good_indicator.apply(abnormal_fix)

data_bad_indicator = data_bad_indicator.apply(abnormal_fix)


#添加排序和好坏标识，构建入模data

data_good_indicator['bad_flag'] = raw_data_good['bad_flag']


data_bad_indicator['bad_flag'] = raw_data_bad['bad_flag']

data = pd.concat([data_good_indicator,data_bad_indicator])

data = data.reset_index(drop=True)
```

## 1.6AR值分析

```python
"""

定义 AR值计算函数

"""

def AR_cal(predictions, labels, cut_point=10):
    
    import numpy as np
    
    sample_size = len(labels)
    
    bad_label_size = len([i for i in labels if i == 1])
    
    #score_thres = np.linspace(1, 0, cut_point)
    
    score_thres = np.linspace( max(predictions), min(predictions), cut_point)
    
    
    x_list = []
    
    y_list = []
    
    x_list.append(0)
    
    y_list.append(0)
    
    for thres in score_thres:
        
        # 阈值以上的样本数 / 总样本数
        x = len([i for i in predictions if i >= thres])       
        
        x_list.append(x / sample_size)
              
        # 阈值以上的样本真实为坏客户的样本数 / 总坏客户样本数
        y = len([(i, j) for i, j in zip(predictions, labels) if i >= thres and j == 1])
              
        y_list.append(y / bad_label_size)
        
    # 计算AR值
    # 实际曲线下面积
    actual_area = np.trapz(y_list, x_list) - 1 * 1 / 2
    
    best_area = 1 * 1 / 2 - 1 * (bad_label_size / sample_size)/2 
   
    ar_value = actual_area / best_area

    return ar_value
```

```python
"""

分析每个变量的AR值,若正向相关某指标值越大，评级越好，AR值小于0，反之AR值大于0

"""

AR_lis = []

for each in data.columns.values:
    
    col = ['指标名称','AR值']
    
    df = pd.DataFrame(columns=col)
    
    df.loc[0,'指标名称'] = each 
    
    #将每个指标的值转换到0-1之间，由于地方政府数据不存在负值，除以最大值即可
    df.loc[0,'AR值'] = AR_cal(predictions=data[each]
                             ,labels=data['bad_flag'], cut_point=1000)
    
    AR_lis.append(df)
    
AR_table = pd.concat(AR_lis)

AR_table


"""AR值大于1的是由于该指标只有唯一值或为空"""
```

## 2.1Logit转换

```python
"""

统一数据转换：Logit转换, X'= 1/(1+exp(a+b*x))

"""

#logit转换函数
def logit_transfer(x,alpha) :
    
    import math
    
    #计算alpha分位数
    P_alpha = np.percentile(a=x,q=100*alpha)
    
    #计算1-alpha分位数
    P_1_aplha = np.percentile(a=x,q=100-alpha*100)
    
    #计算b
    b = 2*math.log(alpha/(1-alpha))/(P_1_aplha-P_alpha)
    
    #计算a
    a = math.log(alpha/(1-alpha))-b*P_1_aplha
    
    x = x.apply(lambda x : 1/(1+math.exp(a+b*x)))
    
    return x



data_transfer = data.apply(logit_transfer,args=(0.01,))



```

## 2.2标准化处理

```python
"""

标准化处理, X'= (X-mean(X))/std(X)

"""

#标准化函数,统一量纲
def standar_transfer(x):
    
    Mean = np.nanmean(x)
    
    Stv = np.nanstd(x)
    
    x = x.apply(lambda x :  (x-Mean)/Stv)
    
    return x


data_transfer = data_transfer.apply(standar_transfer)

```

```python
"""

多重共线性分析，删除VIF值大于5的变量：

定义自动删除函数

"""

import numpy as np
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

## 每轮循环中计算各个变量的VIF，并删除VIF>threshold 的变量
def vif_del(X, thres=10.0):
    
    col = list(range(X.shape[1]))
    
    
    dropped = True
    
    while dropped:
        
        dropped = False
        
        X['c'] = 1
        
        vif = [variance_inflation_factor(X.iloc[:,col].values, ix)
               
               for ix in range(X.iloc[:,col].shape[1])]
        
        
        #print(vif)
        
        maxvif = max(vif)
        
        maxix = vif.index(maxvif)
        
        if maxvif > thres:
            
            print('delete=',X.columns[col[maxix]],'  ', 'vif=',maxvif )
            
            del col[maxix]
            
            dropped = True
            
    print('Remain Variables:', list(X.columns[col]))
    
    print('VIF:', vif)
    
    return list(X.columns[col]) 
```

## 3.2逻辑回归

```python
"""

Somer's D计算函数

计算 数列x 与 数列y 的 Somers' D。
    
计算公式：SomersD(x,y) = tau(x,y) / tau(y,y)

当 数列y 中数值仅为0或1时，AR(x,y) = SomersD(x,y)
    
"""

def somersd(x, y):

    from collections import Counter
    
    # 移除空值
    df = pd.DataFrame({'x': x, 'y': y}).dropna().reset_index()
    
    x = df['x']
    
    y = df['y']
    
    # 检测输入数列长度是否为0
    l = len(y)
    
    if l == 0:
        
        return 0
    
    # 计算 Somers' D
    tau_xy = kendalltau(x, y, use_ties=False)[0] 
    
    tau_yy = 1 - sum([x*(x-1)/2 for x in Counter(y).values()]) / (l*(l-1)/2)
    
    return tau_xy / tau_yy
```

```python
"""

定义 模型结果AR值计算函数

"""

def model_AR_cal(predictions, labels, cut_point=100):
    
    import numpy as np
    
    sample_size = len(labels)
    
    bad_label_size = len([i for i in labels if i == 1])
    
    score_thres = np.linspace(1, 0, cut_point)
    
    #score_thres = np.linspace(0, 1, cut_point)
    
    
    x_list = []
    
    y_list = []
    
    x_list.append(0)
    
    y_list.append(0)
    
    for thres in score_thres:
        
        # 阈值以上的样本数 / 总样本数
        x = len([i for i in predictions if i >= thres])
        
        x_list.append(x / sample_size)
        
        # 阈值以上的样本真实为坏客户的样本数 / 总坏客户样本数
        y = len([(i, j) for i, j in zip(predictions, labels) if i >= thres and j == 1])
        
        y_list.append(y / bad_label_size)


    # 计算AR值
    # 实际曲线下面积
    actual_area = np.trapz(y_list, x_list) - 1 * 1 / 2
    
    best_area = 1 * 1 / 2 - 1 * (bad_label_size / sample_size)/2 
    
    ar_value = actual_area / best_area

    return ar_value
```

```python
"""

对所有的组合方式进行逻辑回归

"""


#遍历-逻辑回归函数

def logistc_all(input_table,group,iter_number,ind_lis,dic,method='ncg') :

    final_lis = []
    
    #对组合列表中的每一个组合进行逻辑回归，记录逻辑回归系数和表现结果
    for each in group:

        temp = each.split('-')[:-1]
        
        #记录组合涉及的指标
        way_lists =[dic[i] for i in temp]
        
        way = '-'.join(way_lists)

        model_lis = []
        

        for i in range(0,len(temp)) : 

            model_lis.append(dic[temp[i]])
            
        #截取组合对应的指标作为自变量

        X = input_table[model_lis]
        
        #截取好坏标识作为因变量
        Y = input_table['bad_flag']
        
        
        #调用statsmodels库中的logit回归
        #maxiter为设置的迭代次数
        import statsmodels.api as sm
        
        #在最前方添加常数列，常数列取值均为1
        X.insert(loc=0,column='const',value=1.0)
        
        
        logit = sm.Logit(Y,X)
        
        #默认采用牛顿-共轭梯度法作为损失函数的优化方法
        RESULT = logit.fit(method=method,maxiter=iter_number)
        
        #采用L-BFGS(Limited-memory BFGS)作为损失函数的优化方法
        #RESULT = logit.fit(method='lbfgs',maxiter=iter_number)
        
        column = ind_lis + [i+'_回归系数卡方P值' for i in ind_lis]
        
        record_table = pd.DataFrame(columns=column)
        
        
        for j in range(0,len(model_lis)): 
            
            #wald检验的卡方P值
            
            wald_test = RESULT.wald_test_terms().table
                     
            #记录各个自变量的系数
            record_table.loc[0,model_lis[j]] = RESULT.params[model_lis[j]]
            
            record_table.loc[0,model_lis[j]+'_回归系数卡方P值'] = RESULT.wald_test_terms().table.loc[model_lis[j],'pvalue']
            
        
        #记录F检验的P值
        A = np.identity(len(RESULT.params))
        
        A = A[1:,:]

        f = RESULT.f_test(A)
        
        #记录F检验_p值
        record_table.loc[0,'F检验_p值'] = float(f.pvalue)
        
        #记录F值
        record_table.loc[0,'F值'] = float(f.fvalue)
        
        #记录McFadden’s pseudo-R-squared
        record_table.loc[0,'类R方_pseudoR方'] = RESULT.prsquared
        
        #记录 似然比 P 值
        record_table.loc[0,'似然比检验P值'] = RESULT.llr_pvalue
        
        #记录 AR 值        
        record_table.loc[0,'AR值'] = model_AR_cal(predictions=RESULT.predict(X), labels=Y, cut_point=1000)
        
        #记录Somer's D
        record_table.loc[0,'somersd值'] = somersd(x=RESULT.predict(X), y=Y)
        
        #记录相关性系数
        record_table.loc[0,'预测结果与原结果相关性系数'] = np.corrcoef(RESULT.predict(X),Y)[0][1]
        
        #记录该模型的指标个数
        record_table.loc[0,'指标个数'] = len(temp)
        
        
        #记录组合方式
        record_table.loc[0,'组合方式'] = way
        
        final_lis.append(record_table)
        
    table = pd.concat(final_lis)
    
    return table



#1.input_table为数据处理及转换后的原始数据集
#2.group为所有的遍历组合方式表
#3.iter_number为设置的迭代次数
#4.ind_lis为纳入多因素分析的指标清单
#5.dic为基于指标清单构建的编号-指标对应字典
final_table = logistc_all(input_table = data_multi_var,group=all_group,iter_number=10000,ind_lis=indicator_lis_new,dic=indicator_dic)



#存储结果
#final_table.to_excel(os.getcwd() + '\\逻辑回归遍历结果.xlsx')
```

```python
"""

定义 Woe值计算函数

"""

def Woe_cal(indictor_table, labels, cut_point=11):
    
    import numpy as np
     
    bad_label_size = len([i for i in labels if i == 1])
    
    good_sample_size = len(labels) - bad_label_size
    
    
    box_thres = np.linspace( min(indictor_table), max(indictor_table), cut_point)
    
    
    Woe_lis = []
    
    
    for k in range(0,len(box_thres)-1):
        
       
        if k+1 != len(box_thres)-1 :
            
            # 档内总样本数
            x = len([i for i in indictor_table if i >= box_thres[k] and i < box_thres[k+1]])
            # 档内违约样本数
            y = len([(i, j) for i, j in zip(indictor_table, labels) if i >= box_thres[k] and i < box_thres[k+1] and j == 1])
            
        else :
            # 档内总样本数
            x = len([i for i in indictor_table if i >= box_thres[k] and i <= box_thres[k+1]])
            # 档内违约样本数
            y = len([(i, j) for i, j in zip(indictor_table, labels) if i >= box_thres[k] and i <= box_thres[k+1] and j == 1])
        

        Woe = np.log((y/bad_label_size)/((x-y)/good_sample_size))
        
        print(k,len(box_thres)-1)
        
        Woe_lis.append(Woe)


    return Woe_lis
```
